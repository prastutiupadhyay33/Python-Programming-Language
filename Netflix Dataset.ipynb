```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
year_count = df['Year'].value_counts().head(10)

plt.bar(year_count.index, year_count.values)
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.title("Bar Chart of Top 10 Release Years on Netflix")
plt.show()
```

```python
plt.hist(df['Year'])
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.title("Histogram of Netflix Release Years")
plt.show()
```

```python
top_years = df['Year'].value_counts().head(5)

plt.pie(top_years.values, labels=top_years.index, autopct='%1.1f%%')
plt.title("Pie Chart of Top 5 Release Years")
plt.show()
```

```python
year_trend = df['Year'].value_counts().sort_index()

plt.plot(year_trend.index, year_trend.values)
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.title("Line Chart of Netflix Releases Over Years")
plt.show()
```

```python
plt.scatter(df['Year'], df.index)
plt.xlabel("Release Year")
plt.ylabel("Index")
plt.title("Scatter Plot of Release Year")
plt.show()
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.hist(df['Year'])
plt.title("Histogram")

plt.subplot(2, 2, 2)
plt.bar(year_count.index, year_count.values)
plt.title("Bar Chart")

plt.subplot(2, 2, 3)
plt.plot(year_trend.index, year_trend.values)
plt.title("Line Chart")

plt.subplot(2, 2, 4)
plt.scatter(df['Year'], df.index)
plt.title("Scatter Plot")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.relplot(x='Duration', y='Year', data=df)
```

```python
sns.scatterplot(x='Duration', y='Year', data=df)
plt.title("Scatter Plot: Duration vs Release Year")
plt.show()
```

```python
year_count = df['Year'].value_counts().sort_index()
sns.lineplot(x=year_count.index, y=year_count.values)
plt.title("Line Plot of Netflix Releases Over Years")
plt.show()
```

```python
sns.histplot(df['Year'])
plt.title("Histplot of Release Year")
plt.show()
```

```python
sns.kdeplot(df['Year'])
plt.title("KDE Plot of Release Year")
plt.show()
```

```python
sns.boxplot(y='Duration', data=df)
plt.title("Boxplot of Duration")
plt.show()
```

```python
sns.violinplot(y='Duration', data=df)
plt.title("Violin Plot of Duration")
plt.show()
```

```python
rating_count = df['Age Rating'].value_counts().reset_index()
rating_count.columns = ['rating', 'count']

sns.barplot(x='rating', y='count', data=rating_count)
plt.title("Barplot of Rating Counts")
plt.xticks(rotation=45)
plt.show()
```

```python

sns.countplot(x='Category', data=df)
plt.title("Countplot of Content Type")
plt.show()

```

```python

numeric_df = df.select_dtypes(include='number')


corr = numeric_df.corr()


plt.figure()
plt.imshow(corr)
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()

```

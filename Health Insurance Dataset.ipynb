```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.bar(df.index[:10], df['charges'][:10])
plt.xlabel("Policy Index")
plt.ylabel("Charges")
plt.title("Bar Chart of First 10 Insurance Charges")
plt.show()
```

```python
plt.hist(df['charges'])
plt.xlabel("Charges")
plt.ylabel("Frequency")
plt.title("Histogram of Insurance Charges")
plt.show()
```

```python
charge_bins = pd.cut(df['charges'], bins=3)
charge_count = charge_bins.value_counts()

plt.pie(charge_count, labels=charge_count.index, autopct='%1.1f%%')
plt.title("Pie Chart of Insurance Charges Distribution")
plt.show()
```

```python
plt.plot(df.index, df['charges'])
plt.xlabel("Policy Index")
plt.ylabel("Charges")
plt.title("Line Chart of Insurance Charges")
plt.show()
```

```python
plt.scatter(df['bmi'], df['charges'])
plt.xlabel("BMI")
plt.ylabel("Charges")
plt.title("Scatter Plot: BMI vs Insurance Charges")
plt.show()
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.hist(df['charges'])
plt.title("Histogram")

plt.subplot(2, 2, 2)
plt.scatter(df['bmi'], df['charges'])
plt.title("BMI vs Charges")

plt.subplot(2, 2, 3)
plt.plot(df.index, df['charges'])
plt.title("Line Chart")

plt.subplot(2, 2, 4)
plt.bar(df.index[:10], df['charges'][:10])
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
```

```python
sns.relplot(x='bmi', y='charges', data=df)
```

```python
sns.scatterplot(x='bmi', y='charges', data=df)
plt.title("Scatter Plot: BMI vs Charges")
plt.show()
```

```python
sns.lineplot(x=df.index, y='charges', data=df)
plt.title("Line Plot of Insurance Charges")
plt.show()
```

```python
sns.histplot(df['charges'])
plt.title("Histplot of Insurance Charges")
plt.show()
```

```python
sns.kdeplot(df['charges'])
plt.title("KDE Plot of Insurance Charges")
plt.show()
```

```python
sns.boxplot(y='charges', data=df)
plt.title("Boxplot of Insurance Charges")
plt.show()
```

```python
sns.violinplot(y='charges', data=df)
plt.title("Violin Plot of Insurance Charges")
plt.show()
```

```python
avg_charges = df.groupby('children')['charges'].mean().reset_index()

sns.barplot(x='children', y='charges', data=avg_charges)
plt.title("Barplot of Average Charges by Number of Children")
plt.show()
```

```python
sns.countplot(x='children', data=df)
plt.title("Countplot of Number of Children")
plt.show()
```

```python
numeric_df = df[['age', 'bmi', 'children', 'charges']]

sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

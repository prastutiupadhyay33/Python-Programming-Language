```python
# Program to perform all arithmetic operations

# Taking user input for two numbers
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Performing operations
print("\nResults of Arithmetic Operations:")
print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)
print("Division:", num1 / num2)
print("Floor Division:", num1 // num2)
print("Modulus:", num1 % num2)
print("Exponentiation:", num1 ** num2)
```

```python
# Swapping list elements using a temporary variable

my_list = [10, 20, 30, 40]
print("Original list:", my_list)

# Swap first and last
temp = my_list[1]
my_list[1] = my_list[3]
my_list[3] = temp

print("List after swapping using temporary variable:", my_list)

# Swapping list elements without using a temporary variable

my_list = [10, 20, 30, 40]
print("Original list:", my_list)

my_list[1], my_list[3] = my_list[3], my_list[1]

print("List after swapping without temporary variable:", my_list)
```

```python
# Creating a dictionary and performing common operations

# Create dictionary
student = {
    "name": "Alice",
    "age": 21,
    "course": "Computer Science"
}
print("Original dictionary:", student)

# Adding elements
student["grade"] = "A"
print("After adding element:", student)
student.update({'address':'Rajiv Nagar'} ) # Update Method
print('After using update:',student)

# Removing an element
removed_item = student.pop("age")  # Pop Method
print("After removing 'age':", student) 
del student['address']  # Delete Method
print('After using delete method:',student)

# Copying dictionary
copy_dict = student.copy()  # Copy Method
print("Copy of dictionary:", copy_dict)
dict_copy=dict(student)  # Dictionary Method
print('Copied dictionary:', dict_copy)

# Creating nested dictionary
nested_dict = {
    "student1": {"name": "Alice", "age": 21},
    "student2": {"name": "Bob", "age": 22}
}
print("Nested Dictionary:", nested_dict)
```

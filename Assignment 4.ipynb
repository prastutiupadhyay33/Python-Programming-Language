```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Relplot – Sales vs Profit
sns.relplot(x='Sales', y='Profit', data=df)
plt.title("Relational Plot: Sales vs Profit")
plt.show()
```

```python
# Scatterplot – Sales vs Profit
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Scatter Plot: Sales vs Profit")
plt.show()
```

```python
# Histplot – Sales Distribution
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title("Histogram: Sales Distribution")
plt.show()
```

```python
# KDE Plot – Profit Distribution
sns.kdeplot(df['Profit'], fill=True)
plt.title("KDE Plot: Profit Distribution")
plt.show()
```

```python
# Boxplot – Sales (Outlier Detection)
sns.boxplot(y='Sales', data=df)
plt.title("Boxplot: Sales")
plt.show()
```

```python
# Violin Plot – Profit by Category
sns.violinplot(x='Category', y='Profit', data=df)
plt.title("Violin Plot: Profit by Category")
plt.show()
```

```python
# Barplot – Average Sales by Category
sns.barplot(x='Category', y='Sales', data=df)
plt.title("Bar Plot: Average Sales by Category")
plt.show()
```

```python
# Lineplot-Sales trend in Superstore
df = df.reset_index()

sns.lineplot(x="index", y="Sales", data=df)

plt.xlabel("Index")
plt.ylabel("Sales")
plt.title("Sales Trend in Superstore Dataset")
plt.show()
```

```python
# Countplot – Orders by Segment
sns.countplot(x='Segment', data=df)
plt.title("Count Plot: Orders by Segment")
plt.show()
```

```python
# Heatmap 
plt.figure(figsize=(7,5))
sns.heatmap(df[['Sales','Profit','Quantity','Discount']].corr(),
            annot=True, cmap='coolwarm')
plt.title("Heatmap:Correlation Matrix")
plt.show()
```

```python
# Pairplot
sns.pairplot(df[['Sales','Profit','Discount']])
plt.show()
```

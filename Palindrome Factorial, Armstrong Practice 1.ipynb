```python
cars=['Ford','Maruti','Audi']
for i in cars:
    print(i)
```

```python
cars=['Ford','Maruti','Audi']
for i in cars:
    if(i=='Audi'):
      break
    print(i)
```

```python
cars=['Ford','Maruti','Audi']
for i in cars:
    print(i)
    if(i=='Audi'):
      break
```

```python
cars=['Ford','Maruti','Audi']
for i in cars:
    if(i=='Audi'):
      continue
    print(i)
```

```python
# Nested For Loop
adj=['red','big','tasty']
fruits=['apple','banana','cherry']
for x in adj:
   for y in fruits:
     print(x,y)
```

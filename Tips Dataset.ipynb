```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Line chart
plt.figure()
plt.plot(df['total_bill'], df['tip'])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Line Chart: Total Bill vs Tip")
plt.show()
```

```python
#Scatter plot
plt.figure()
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

```python
#Bar Chart
avg_tip = df.groupby('day')['tip'].mean()

plt.figure()
plt.bar(avg_tip.index, avg_tip.values)
plt.xlabel("Day")
plt.ylabel("Average Tip")
plt.title("Bar Chart: Average Tip by Day")
plt.show()
```

```python
#Histogram
plt.figure()
plt.hist(df['total_bill'])
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("Histogram: Total Bill Distribution")
plt.show()
```

```python

plt.figure(figsize=(15, 8))

#  Line Plot
plt.subplot(2, 3, 1)
plt.plot(df['total_bill'], df['tip'])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Line Plot")

#  Scatter Plot
plt.subplot(2, 3, 2)
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot")

#  Bar Chart – Avg Tip by Day
avg_tip = df.groupby('day')['tip'].mean()
plt.subplot(2, 3, 3)
plt.bar(avg_tip.index, avg_tip.values)
plt.xlabel("Day")
plt.ylabel("Avg Tip")
plt.title("Bar Chart")

# 4️⃣ Histogram
plt.subplot(2, 3, 4)
plt.hist(df['total_bill'])
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("Histogram")


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#bar chart
plt.bar(df.index[:10], df['G3'][:10])
plt.xlabel("Student Index")
plt.ylabel("Final Grade (G3)")
plt.title("Bar Chart of First 10 Students Final Grades")
plt.show()

```

```python
#Histogram
plt.hist(df['G3'])
plt.xlabel("Final Grade (G3)")
plt.ylabel("Frequency")
plt.title("Histogram of Final Grades")
plt.show()


```

```python
#Pie chart
grade_bins = pd.cut(df['G3'], bins=3)
grade_count = grade_bins.value_counts()

plt.pie(grade_count, labels=grade_count.index, autopct='%1.1f%%')
plt.title("Pie Chart of Grade Distribution")
plt.show()
```

```python
#Line chart
plt.plot(df.index, df['G3'])
plt.xlabel("Student Index")
plt.ylabel("Final Grade (G3)")
plt.title("Line Chart of Students Final Grades")
plt.show()
```

```python
#Scatter plot
plt.scatter(df['studytime'], df['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.title("Scatter Plot: Study Time vs Final Grade")
plt.show()
```

```python
#Subplot
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.hist(df['G3'])
plt.title("Histogram of Grades")

plt.subplot(2, 2, 2)
plt.scatter(df['studytime'], df['G3'])
plt.title("Study Time vs Grade")

plt.subplot(2, 2, 3)
plt.plot(df.index, df['G3'])
plt.title("Line Chart of Grades")

plt.subplot(2, 2, 4)
plt.bar(df.index[:10], df['G3'][:10])
plt.title("Bar Chart of Grades")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
#Relplot
sns.relplot(x='studytime', y='G3', data=df)
```

```python
#Scatterplot
sns.scatterplot(x='studytime', y='G3', data=df)
plt.title("Scatter Plot: Study Time vs Final Grade")
plt.show()
```

```python
#lineplot
sns.lineplot(x=df.index, y='G3', data=df)
plt.title("Line Plot of Students Final Grades")
plt.show()
```

```python
#Distplot
sns.distplot(df['G3'])
plt.title("Distplot of Final Grades")
plt.show()
```

```python
#Histplot
sns.histplot(df['G3'])
plt.title("Histplot of Final Grades")
plt.show()
```

```python
#KDE plot
sns.kdeplot(df['G3'])
plt.title("KDE Plot of Final Grades")
plt.show()
```

```python
#Boxplot
sns.boxplot(y='G3', data=df)
plt.title("Boxplot of Final Grades")
plt.show()
```

```python
#Violin plot
sns.violinplot(y='G3', data=df)
plt.title("Violin Plot of Final Grades")
plt.show()
```

```python
#Barplot
sns.barplot(x=df.index[:10], y=df['G3'][:10])
plt.title("Barplot of First 10 Students Final Grades")
plt.show()
```

```python
#Countplot
grade_category = pd.cut(df['G3'], bins=3)
sns.countplot(x=grade_category)
plt.title("Countplot of Grade Categories")
plt.show()
```

```python
#Heatmap
numeric_df = df.select_dtypes(include=['int64', 'float64'])

sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Pairplot
numeric_df = df.select_dtypes(include=['int64', 'float64'])

sns.pairplot(numeric_df[['studytime', 'absences', 'age', 'G3']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Bar chart

plt.bar(df.index[:10], df['Price'][:10])
plt.xlabel("House Index")
plt.ylabel("Price")
plt.title("Bar Chart of First 10 House Prices")
plt.show()
```

```python
#Histogram

plt.hist(df['Price'])
plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.title("Histogram of House Prices")
plt.show()
```

```python
#Pie chart
price_bins = pd.cut(df['Price'], bins=3)
price_count = price_bins.value_counts()

plt.pie(price_count, labels=price_count.index, autopct='%1.1f%%')
plt.title("Pie Chart of House Price Distribution")
plt.show()
```

```python
#Line Chart

plt.plot(df.index, df['Price'])
plt.xlabel("Index")
plt.ylabel("Price")
plt.title("Line Chart of House Prices")
plt.show()
```

```python
#Scatter plot

plt.scatter(df['Area'], df['Price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Scatter Plot (Area vs Price)")
plt.show()
```

```python
#Subplot

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.hist(df['Price'])
plt.title("Histogram")

plt.subplot(2, 2, 2)
plt.scatter(df['Area'], df['Price'])
plt.title("Scatter Plot")

plt.subplot(2, 2, 3)
plt.plot(df.index, df['Price'])
plt.title("Line Chart")

plt.subplot(2, 2, 4)
plt.bar(df.index[:10], df['Price'][:10])
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns

```

```python
#Relplot

sns.relplot(x='Area', y='Price', data=df)
```

```python
#Scatter plot

sns.scatterplot(x='Area', y='Price', data=df)
plt.title("Scatter Plot: Area vs Price")
plt.show()
```

```python
#Lineplot

sns.lineplot(x=df.index, y='Price', data=df)
plt.title("Line Plot of House Prices")
plt.show()
```

```python
#Distplot


sns.distplot(df['Price'])
plt.title("Distplot of House Prices")
plt.show()


```

```python
#Histplot

sns.histplot(df['Price'])
plt.title("Histplot of House Prices")
plt.show()
```

```python
#KDE plot

sns.kdeplot(df['Price'])
plt.title("KDE Plot of House Prices")
plt.show()
```

```python
#Boxplot

sns.boxplot(y='Price', data=df)
plt.title("Boxplot of House Prices")
plt.show()
```

```python
#Violin plot

sns.violinplot(y='Price', data=df)
plt.title("Violin Plot of House Prices")
plt.show()
```

```python
#Bar plot

sns.barplot(x=df.index[:10], y=df['Price'][:10])
plt.title("Barplot of First 10 House Prices")
plt.show()
```

```python
#Countplot

price_category = pd.cut(df['Price'], bins=3)
sns.countplot(x=price_category)
plt.title("Countplot of Price Categories")
plt.show()
```

```python
#Heatmap
numeric_df = df.select_dtypes(include=['int64', 'float64'])

corr = numeric_df.corr()


sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()

```

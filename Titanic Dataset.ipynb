```python
#Creating different plots
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Bar Chart – Survival Count
survival_counts = df['Survived'].value_counts()

plt.bar(survival_counts.index, survival_counts.values)
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.xlabel("Survival Status")
plt.ylabel("Number of Passengers")
plt.title("Survival Count")
plt.show()
```

```python
#Histogram – Age Distribution
plt.hist(df['Age'].dropna(), bins=10)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
#Pie Chart – Passenger Class Distribution
class_counts = df['Pclass'].value_counts()

plt.pie(class_counts.values, labels=class_counts.index, autopct='%1.1f%%')
plt.title("Passenger Class Distribution")
plt.show()
```

```python
#Line Chart – Survival Rate by Class
survival_rate = df.groupby('Pclass')['Survived'].mean()

plt.plot(survival_rate.index, survival_rate.values, marker='o')
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
# Subplot 
# plot 1: Passenger Class count
plt.subplot(1, 2, 1)
df['Pclass'].value_counts().sort_index().plot()
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.title("Passenger Class Distribution")

# plot 2: Survival count
plt.subplot(1, 2, 2)
df['Survived'].value_counts().plot()
plt.xlabel("Survival (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.title("Survival Distribution")

plt.show()
```

```python
#To check Outliers
#Scatter Plot – Age vs Fare
plt.scatter(df['Age'], df['Fare'])
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age vs Fare")
plt.show()
```

```python
#Boxplot for Age

import matplotlib.pyplot as plt

plt.boxplot(df['Age'].dropna())
plt.title("Boxplot of Age")
plt.ylabel("Age")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.bar(df.index[:10], df['quality'][:10])
plt.xlabel("Wine Index")
plt.ylabel("Quality")
plt.title("Bar Chart of First 10 Wine Qualities")
plt.show()
```

```python
plt.hist(df['quality'])
plt.xlabel("Quality")
plt.ylabel("Frequency")
plt.title("Histogram of Wine Quality")
plt.show()
```

```python
quality_bins = pd.cut(df['quality'], bins=3)
quality_count = quality_bins.value_counts()

plt.pie(quality_count, labels=quality_count.index, autopct='%1.1f%%')
plt.title("Pie Chart of Wine Quality Distribution")
plt.show()
```

```python
plt.plot(df.index, df['quality'])
plt.xlabel("Wine Index")
plt.ylabel("Quality")
plt.title("Line Chart of Wine Quality")
plt.show()
```

```python
plt.scatter(df['alcohol'], df['quality'])
plt.xlabel("Alcohol")
plt.ylabel("Quality")
plt.title("Scatter Plot: Alcohol vs Quality")
plt.show()
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.hist(df['quality'])
plt.title("Histogram of Quality")

plt.subplot(2, 2, 2)
plt.scatter(df['alcohol'], df['quality'])
plt.title("Alcohol vs Quality")

plt.subplot(2, 2, 3)
plt.plot(df.index, df['quality'])
plt.title("Line Chart of Quality")

plt.subplot(2, 2, 4)
plt.bar(df.index[:10], df['quality'][:10])
plt.title("Bar Chart of Quality")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.relplot(x='alcohol', y='quality', data=df)
```

```python
sns.scatterplot(x='alcohol', y='quality', data=df)
plt.title("Scatter Plot: Alcohol vs Quality")
plt.show()
```

```python
sns.lineplot(x=df.index, y='quality', data=df)
plt.title("Line Plot of Wine Quality")
plt.show()
```

```python
sns.histplot(df['quality'])
plt.title("Histplot of Wine Quality")
plt.show()
```

```python
sns.kdeplot(df['quality'])
plt.title("KDE Plot of Wine Quality")
plt.show()
```

```python
sns.boxplot(y='quality', data=df)
plt.title("Boxplot of Wine Quality")
plt.show()
```

```python
sns.violinplot(y='quality', data=df)
plt.title("Violin Plot of Wine Quality")
plt.show()
```

```python
sns.barplot(x=df.index[:10], y=df['quality'][:10])
plt.title("Barplot of First 10 Wines")
plt.show()
```

```python
quality_category = pd.cut(df['quality'], bins=3)
sns.countplot(x=quality_category)
plt.title("Countplot of Wine Quality Categories")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
